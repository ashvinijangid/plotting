```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
type_1 = np.random.random((100,2)) * 5
type_2 = np.random.random((100,2)) * 6
type_3 = np.random.random((100,2)) * 7
```

```python
plt.scatter(type_1[:,0],type_1[:,1])
```

```python
fig = plt.figure(figsize=(20,8),dpi=150)
plt.scatter(type_1[:,0],type_1[:,1], s=100, c='b', alpha=0.6, edgecolors='k')
plt.scatter(type_2[:,0],type_2[:,1], s=100, c='r', alpha=0.6, edgecolors='k')
plt.scatter(type_3[:,0],type_3[:,1], s=100, c='g', alpha=0.6, edgecolors='k')
```
